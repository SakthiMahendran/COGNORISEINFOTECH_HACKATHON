```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/train.csv')
```

```python
df.info()
```

```python
for col in df.columns:
    first_element = df[col].iloc[0]
    if isinstance(first_element, str) and first_element.startswith('A'):
        unique_elements = df[col].unique()
        print(f"Column '{col}' with unique elements: {unique_elements}")
```

```python
# Get the target column name
target_column = 'Group_no'

# Iterate through each column (excluding the target column)
for col in df.columns:
    if col != target_column:
        # Check if the column contains integer values
        if df[col].dtype == 'int64':
            # Calculate correlation for the current column with the target column
            correlation = df[col].corr(df[target_column])
            
            # Plot the correlation
            plt.figure(figsize=(6, 4))
            plt.scatter(df[col], df[target_column], color='skyblue')
            plt.title(f"Correlation between '{col}' and '{target_column}': {correlation:.2f}")
            plt.xlabel(col)
            plt.ylabel(target_column)
            plt.grid(True)
            plt.tight_layout()
            plt.show()
        else:
            print(f"Skipping column '{col}' as it does not contain integer values.")
```
